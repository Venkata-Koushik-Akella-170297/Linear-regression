# Step 2: Preparing the data for model preparation
- Encoding:
    - converting binary categorical variables to 0's and 1's
    - converting other variables into dummy variables
- Splitting the data into train and test dataset

### Encoding the data:

```python
housing.head()
```

```python
# yes/no variables
varlist = ['mainroad', 'guestroom', 'basement', 'airconditioning',
           'prefarea','hotwaterheating']
housing[varlist] = housing[varlist].apply(lambda x: x.map({'yes': 1,'no': 0}))
housing[varlist].head()
```

#### Dummy variables:

```python
housing.head()
```

Here we don't need three columns, we can represent the furnishing status with just two columns as follows by dropping the `furnished` column -
- `00` will correspond to furnished
- `01` will correspond to unfurnished
- `10` will correspond to semi-furnished

```python
# creating dummy variables for furnishing status
# dropping redundant dummy var
status=pd.get_dummies(housing['furnishingstatus'], drop_first=True)
status.head()
```

```python
# concat dummy df with original 
housing = pd.concat([housing, status], axis=1)
housing.head()
```

```python
housing = housing.drop('furnishingstatus', axis=1)
```

```python
housing.head()
```

### Splitting the data into train test dataset:

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(housing, train_size=0.7, random_state=100)
print(df_train.shape)
print(df_test.shape)
```

### Rescaling the Features 

As you saw in the demonstration for Simple Linear Regression, scaling doesn't impact your model. Here we can see that except for `area`, all the columns have small integer values. So it is extremely important to rescale the variables so that they have a comparable scale. If we don't have comparable scales, then some of the coefficients as obtained by fitting the regression model might be very large or very small as compared to the other coefficients. This might become very annoying at the time of model evaluation. So it is advised to use standardization or normalization so that the units of the coefficients obtained are all on the same scale. As you know, there are two common ways of rescaling:

1. Min-Max scaling 
2. Standardisation (mean-0, sigma-1) 

This time, we will use MinMax scaling.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# instantiate an object
scaler = MinMaxScaler()

# create a list of num vars
numvars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']

#fit the data
df_train[numvars] = scaler.fit_transform(df_train[numvars])
df_train.head()
```

```python
df_train[numvars].describe()
```

- We can look significance of variables and drop variables based on significance only or
- we can also look at VIF score for model

### Checking VIF

Variance Inflation Factor or VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other. It is an extremely important parameter to test our linear model. The formula for calculating `VIF` is:

### $ VIF_i = \frac{1}{1 - {R_i}^2} $

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# creating a df that will contains the names of all the feature variables and their respective VIF values
vif = pd.DataFrame()
vif['features'] = x_train.columns
vif['VIF'] = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

we now have 3 cases
- high p, high vif : remove the variable
- high-low
    - high p , low vif : first drop this first and build a new model
    - low p , high vif
- low p, low vif : keep the variable

```python
x_train.head()
```

```python
x = x_train.drop('semi-furnished', axis=1) # dropping semi-furnished as it has high p-value
```

```python
# add a constant
x_train_sm = sm.add_constant(x)
# creating the model
lr = sm.OLS(y_train, x_train_sm)
# fitting the model
lr_model = lr.fit()
# parameters of the model
lr_model.summary()
```

```python
# Calculate the VIFs again for the new model

vif = pd.DataFrame()
vif['Features'] = x.columns
vif['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
x = x.drop('bedrooms',axis=1)
```

```python
# add a constant
x_train_sm = sm.add_constant(x)
# creating the model
lr = sm.OLS(y_train, x_train_sm)
# fitting the model
lr_model = lr.fit()
# parameters of the model
lr_model.summary()
```

```python
# Calculate the VIFs again for the new model

vif = pd.DataFrame()
vif['Features'] = x.columns
vif['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# Step 4: (a) Residual Analysis

```python
y_train_pred = lr_model.predict(x_train_sm)
```

```python
res = y_train - y_train_pred
plt.figure(figsize=(10,7))
sns.distplot(res)
plt.show()
```

# Step 4: (b) Prediction and Evaluation of the Test set

A point of caution here
we only fit the model on the Training dataset and transform the set data set in the following step

`df_test[numvars] = scaler.transform(df_test[numvars])`

because we cannot try to fit the model on the test data set as it is only meant to test the already fit model

```python
# create a list of num vars
numvars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']

#transfor the test dataset
df_test[numvars] = scaler.transform(df_test[numvars])
df_test.head()
```

```python
df_test.describe()
```

```python
y_test = df_test.pop("price")
x_test = df_test
```

```python
# add a constant to x_test
x_test_sm = sm.add_constant(x_test)
x_test_sm.head()
```

```python
x_test_sm=x_test_sm.drop(['bedrooms', 'semi-furnished'], axis=1)
```

```python
# predictions
y_test_pred = lr_model.predict(x_test_sm)
```

```python
# evaluate
from sklearn.metrics import r2_score
r2_score(y_true = y_test, y_pred = y_test_pred)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_test_pred)
plt.title("y_test vs y_pred")
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.show()
```

the value of r2_score for training set was 0.68 and for the test set was found to be 0.671 which are pretty close to each other and hence are our model is able to reciprocate what it had learned in the training phase.


We can see that the equation of our best fitted line is:

$ price = 0.236  \times  area + 0.202  \times  bathrooms + 0.11 \times stories + 0.05 \times mainroad + 0.04 \times guestroom + 0.0876 \times hotwaterheating + 0.0682 \times airconditioning + 0.0629 \times parking + 0.0637 \times prefarea - 0.0337 \times unfurnished $

Overall we have a decent model, but we also acknowledge that we could do better. 

We have a couple of options:
1. Add new features (bathrooms/bedrooms, area/stories, etc.)
2. Build a non-linear model
